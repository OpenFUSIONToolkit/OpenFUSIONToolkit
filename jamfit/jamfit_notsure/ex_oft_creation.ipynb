# Example OFT CREATION 
This is a notebook can help create oft xml files for specific cases. Requires resistivites, and locations of icoils and vcoils. This is an example notebook

```python
import xml.etree.cElementTree as ET
from build_xml import Vcoil, Icoil, create_oft
```

# Loading in data
Typically I load in data through a text file that is in the format of Coil Name \n R Z \n R Z etc. 

You can have any kind of input you want as long as it can interface with initializing the classes correctly 

```python
data = {}

with open("fcoil_pts_test.txt", "r") as f:
    lines = f.readlines()

current_name = None
for line in lines:
    line = line.strip()
    if not line:
        continue  # skip blank lines

    parts = line.split()
    if len(parts) == 1:
        # It's a name
        current_name = parts[0]
        data[current_name] = {'radius': [], 'Z': []}
    elif len(parts) == 2:
        # It's an r z pair
        r, z = map(float, parts)
        data[current_name]['radius'].append(r)
        data[current_name]['Z'].append(z)
    else:
        raise ValueError(f"Unexpected line format: {line}")
```

```python
coil_list = [Icoil(name, values['radius'], values['Z']) for name, values in data.items()]
resistivites = [1E-4]
create_oft('fcoil.xml', coil_list, [], resistivites) 
```

# Creating a custom coil xml file
I did this in a seperate xml file, but you can join just the same with coil_list. 

```python
data = {} 
with open("custom_coil.txt", "r") as f:
    lines = f.readlines()
name = None 
for line in lines: 
    line = line.strip()
    if not line:
        continue  # skip blank lines
    parts = line.split()
    if len(parts) == 1:
        name = parts[0]
        data[name] = {'xyz': []}
    elif len(parts) == 3:
        x, y, z = map(float, parts)
        data[name]['xyz'].append((x, y, z))
    else:
        raise ValueError(f"Unexpected line format: {line}")
```

```python
coil_list = [Vcoil(name, radius=10, Z=None, xyz=values['xyz'], scale=0, resistivity_per_length=0.01, sens_mask=1, npoints=3, iscustom=True) for name, values in data.items()]
create_oft('custom_coil.xml', [], coil_list, resistivites) 
```
